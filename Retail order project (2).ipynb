```python
import pandas as pd 
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df.head(20)
df['Ship Mode'].unique()
```

Derive new columns discount and price , profit

```python
df['dicount']=df['list price']*df['discount percent']*.01
df
```

```python
df.head()
```

```python
df['sale_price'] = df['list price'] - df['dicount']
df
```

Rename column dicount to discount

```python
df.rename(columns={'dicount': 'Discount'}, inplace=True)
```

```python
df
```

calculating profits

```python
df['profit']=df['sale_price']-df['cost price']
```
